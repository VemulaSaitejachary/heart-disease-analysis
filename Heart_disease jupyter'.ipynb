# Heart Disease

```python
colors = ['red', 'green']
data['target'].value_counts().plot(kind='bar',color=colors)
plt.xticks([1,0],['No-Disease','Disease'])
plt.title('Heart Disease')
plt.show()
```

```python
sns.countplot(x='sex',hue='target',data=data)
plt.title('Heart Disease Caused By Gender')
plt.xticks([1,0],['Male','Female'])
plt.legend(labels=['No Diseases','Diseases'])
plt.show()
```

# Chest pain types

```python
# Typical Angina (TA), 
# Atypical Angina (ATA), 
# Non-Anginal Pain (NAP)
# Asymptomatic (ASY)
```

```python
data.columns
```

```python
sns.countplot(x='cp',data=data)
plt.xlabel('Chest Pain')
plt.xticks([0,1,2,3],['Typical Angina','Atypical Angin','Non-Anginal Pain','Asymptomatic'])
plt.xticks(rotation=75)
plt.show()
```

# Chest Pain Distibution as per Target

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['No Disease','Disease'])
plt.show()
```

# Fasting blood sugar as per our target

```python
# with fbs > 120 is consider diabatics 
sns.countplot(x='fbs',data=data,hue='target')
plt.xlabel('Fasting blood sugar')
plt.legend(labels=['No-Disease','Disease'])
plt.show()
```
